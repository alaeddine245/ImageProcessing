```python
ppm_ops = PpmOperations()
```

```python
chat_rgb = ppm_ops.read('chat.ppm')
```

```python
chat_rgb.show()
```

```python
ppm_ops.draw_histogram(chat_rgb)
```

```python
ppm_ops.threshhold(chat_rgb, (100,140,90), 'OR').show()
```

```python
histogram_red = ppm_ops.histogram(chat_rgb , 0)
histogram_green = ppm_ops.histogram(chat_rgb, 1)
histogram_blue = ppm_ops.histogram(chat_rgb, 2)


otsu_red = ppm_ops.otsu_thresholding(histogram_red)
otsu_green = ppm_ops.otsu_thresholding(histogram_green)
otsu_blue = ppm_ops.otsu_thresholding(histogram_blue)
```

```python
result = ppm_ops.threshhold(chat_rgb, (otsu_red, otsu_green, otsu_blue), 'AND')
```

```python
result.show()
```

```python
ppm_ops.erosion(result, 3).show()
```
